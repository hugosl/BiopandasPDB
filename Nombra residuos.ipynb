```python
import pandas as pd
from biopandas.pdb import PandasPdb


# Cargar el archivo PDB utilizando BioPandas
pdb = PandasPdb().read_pdb('6y2e.pdb')


residues = pdb.df['ATOM'][['residue_name', 'residue_number']]

# Crear un conjunto para almacenar residuos únicos
unique_residues = set()

# Recorrer los residuos e imprimir su código y número si no están en el conjunto de residuos únicos
for index, row in residues.iterrows():
    residue_code = row['residue_name']
    residue_number = row['residue_number']
    
    # Verificar si el residuo ya se ha registrado
    if (residue_code, residue_number) in unique_residues:
        continue  # Saltar a la siguiente iteración si ya está registrado
    
    # Agregar el residuo al conjunto de residuos únicos
    unique_residues.add((residue_code, residue_number))
    
    # Imprimir el código y número del residuo
    print(f"Código: {residue_code}, Número: {residue_number}")
```

```python
import pandas as pd
from biopandas.pdb import PandasPdb

# Cargar el archivo PDB utilizando BioPandas
pdb = PandasPdb().read_pdb('6y2e.pdb')

residues = pdb.df['ATOM'][['residue_name', 'residue_number']]

# Crear un diccionario para mapear números de residuos a códigos de residuos
residue_mapping = {}

# Recorrer los residuos y agregarlos al diccionario si no están presentes
for index, row in residues.iterrows():
    residue_code = row['residue_name']
    residue_number = str(row['residue_number'])  # Convertir el número de residuo a str
    residue_mapping[residue_number] = residue_code

# Solicitar al usuario ingresar el número de residuo
residue_number_input = input("Ingrese el número de residuo: ")

# Buscar el código de residuo correspondiente al número ingresado de forma flexible
residue_code_result = residue_mapping.get(residue_number_input)

# Verificar si se encontró el código de residuo para el número ingresado
if residue_code_result:
    print(f"El código de residuo para el número {residue_number_input} es: {residue_code_result}")
else:
    print(f"No se encontró un código de residuo para el número {residue_number_input}")
```
